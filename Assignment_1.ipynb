### 3. AND / OR Gates Using Perceptron

```python
# y=step(w1​x1​+w2​x2​+b) perceptron logic
import numpy as np

def perceptron(x, w, b):
    return 1 if np.dot(w, x) + b >= 0 else 0
```

```python
w_and = np.array([1, 1])
b_and = -1.5

for x in [(0,0),(0,1),(1,0),(1,1)]:
    print(x, perceptron(x, w_and, b_and))
# and gate
```

```python
w_or = np.array([1, 1])
b_or = -0.5

for x in [(0,0),(0,1),(1,0),(1,1)]:
    print(x, perceptron(x, w_or, b_or))
# or gate
```

### 4. XOR Problem Using PyTorch Neural Network

```python
# XOR need NN because it is not linearly separable 

X = torch.tensor([[0.,0.],[0.,1.],[1.,0.],[1.,1.]])
y = torch.tensor([[0.],[1.],[1.],[0.]])
# dataset
```

```python
import torch.nn as nn

class XORNet(nn.Module):
    def __init__(self):
        super().__init__()
        self.hidden = nn.Linear(2, 2)
        self.output = nn.Linear(2, 1)

    def forward(self, x):
        x = torch.relu(self.hidden(x))
        return torch.sigmoid(self.output(x))
# defining the NN
```

```python
model = XORNet()
criterion = nn.BCELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.1)

for epoch in range(2000):
    optimizer.zero_grad()
    y_pred = model(X)
    loss = criterion(y_pred, y)
    loss.backward()
    optimizer.step()
# training
```

```python
print(model(X).round())
# results
```

### 5. Simple Neural Network for Regression according to the Diagram

Architecture :

2 Inputs: x1, x2
Hidden Layer: 2 neurons
Output Layer: 1 neuron
Activation: ReLU (hidden), Linear (output)

```python
X = torch.rand((100, 2))
y = 3*X[:,0] + 2*X[:,1] + 1
y = y.view(-1,1)
# synthetic regression data
```

```python
class SimpleRegNet(nn.Module):
    def __init__(self):
        super().__init__()
        self.h1 = nn.Linear(2, 2)   # input → hidden
        self.out = nn.Linear(2, 1) # hidden → output

    def forward(self, x):
        z1 = torch.relu(self.h1(x))
        return self.out(z1)
```

```python
model = SimpleRegNet()
criterion = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.05)

for epoch in range(1000):
    optimizer.zero_grad()
    y_pred = model(X)
    loss = criterion(y_pred, y)
    loss.backward()
    optimizer.step()

    if epoch % 100 == 0:
        print(f"Epoch {epoch}, Loss: {loss.item():.4f}")
```

```python
test = torch.tensor([[0.5, 0.8]])
print(model(test))  
# prediction
```
